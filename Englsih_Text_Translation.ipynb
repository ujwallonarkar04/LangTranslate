```python
import pandas as pd
import os
import time
from googletrans import Translator
```

```python
df=pd.read_excel("Order Export.xls")
```

```python
df.head()
```

```python
df.info()
```

```python
translator = Translator()
```

```python
def translate_to_english(text):
    try:
        if pd.notnull(text):
            translation = translator.translate(str(text), src='zh-cn', dest='en')
            return translation.text
        else:
            return ''
    except Exception as e:
        print(f"Translation error: {e}")
        return ''
```

```python
translated_columns = [translate_to_english(col) for col in df.columns]

df.columns = translated_columns
```

```python
start_time = time.time()

for column in df.columns:
    df[column] = df[column].apply(translate_to_english)

end_time = time.time()


```

```python
time_taken = end_time - start_time
print(f"Time taken to translate and save: {time_taken} seconds")
```
